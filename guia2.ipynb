```python
from functools import reduce 
import random 
```

```python
x = lambda x: x**2 + 2*x + 1
print(x(2))
print(x(4))
print(x(6))
```

```python
velocidades = [10, 20, 30, 40, 50] 
m=list(map(lambda x: x*0.2777, velocidades)) 
print (m)
```

```python
valores = [2,5,8,11,14,17,20]
pares=list(filter(lambda x:x % 2 == 0, valores)) 
print (pares)
```

```python
numeros =[2,3,4,5]
resultado = reduce(lambda x,y: x+y, numeros)
print (resultado)
```

```python
data = [12,15,22,35,41,8,3,55,62,90]
resultado = reduce(lambda x,y: x+y, 
map(lambda x: x*2, 
filter(lambda x: x>=10, data)))
print(resultado)
```

```python
data = [random.randint(0,100) for _ in range(1000000)] 
filtrar= filter(lambda x: x > 50, data)
elevar=map(lambda x: x**2, filtrar)
elevados=list(elevar)
promedio=sum(elevados)/len(elevados)
print(promedio)
```

```python
sensores = [random.uniform(10,80) for _ in range(500000)] 
filtrar=list(filter(lambda x: 10 <= x <= 70, sensores)) 
kelvin = map(lambda x: x + 273.15, filtrar)
combertidos=list(kelvin)
promedio=sum(combertidos)/len(combertidos)
print(promedio)
```
